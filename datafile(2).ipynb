#### 4.1 Data Visualization

```python
df.columns = ['Crop', 'Production 2006-07', 'Production 2007-08',
       'Production 2008-09', 'Production 2009-10', 'Production 2010-11',
       'Area 2006-07', 'Area 2007-08', 'Area 2008-09', 'Area 2009-10',
       'Area 2010-11', 'Yield 2006-07', 'Yield 2007-08', 'Yield 2008-09',
       'Yield 2009-10', 'Yield 2010-11']
```

## Area

```python
plt.figure(figsize=(20,7))
sns.lineplot(data=df,x="Crop",y="Area 2006-07",marker='o',label='Area 2006-07')
sns.lineplot(data=df,x="Crop",y="Area 2007-08",marker='o',label='Area 2007-08')
sns.lineplot(data=df,x="Crop",y="Area 2008-09",marker='o',label='Area 2008-09')
sns.lineplot(data=df,x="Crop",y="Area 2009-10",marker='o',label='Area 2009-10')
sns.lineplot(data=df,x="Crop",y="Area 2010-11",marker='o',label='Area 2010-11')
plt.xticks(rotation=90)
plt.xlabel('Crop',fontsize = 12)
plt.ylabel('Area 2006-2011', fontsize = 12)
plt.title('Area-wise Production from 2006 to 2011(each year)' , fontsize = 17)
plt.show()
```

#### Bar-Graph for Area

```python
df['Total_Area'] = df[['Area 2006-07', 'Area 2007-08', 'Area 2008-09', 'Area 2009-10', 'Area 2010-11']].sum(axis=1)

# Sort the DataFrame by total area
df_sorted = df.sort_values(by='Total_Area', ascending=False)

plt.figure(figsize=(15, 7))
sns.barplot(x='Crop', y='Total_Area', data=df_sorted, palette='viridis')
plt.xlabel('Crop', fontsize=12)
plt.ylabel('Total Area (2006-2011)', fontsize=12)
plt.title('Total Area of Crops (2006-2011)', fontsize=17)
plt.xticks(rotation=90)
plt.show()
```

####  Pie-Chart for Area

```python
# Aggregate the area across all years for each crop
df['Total_Area'] = df[['Area 2006-07', 'Area 2007-08', 'Area 2008-09', 'Area 2009-10', 'Area 2010-11']].sum(axis=1)

# Calculate the percentage of total area for each crop
df['Area_Percentage'] = (df['Total_Area'] / df['Total_Area'].sum()) * 100

# Sort the DataFrame by Area_Percentage
df = df.sort_values(by='Area_Percentage', ascending=False)

# Plotting the pie chart
plt.figure(figsize=(10, 10))
plt.pie(df['Area_Percentage'], labels=df['Crop'], startangle=140)
plt.title('Crop Distribution based on Total Area (2006-2011)', fontsize=17)
plt.axis('equal')  
plt.show()
```

####  Histogram for Area

```python
# Aggregate the area across all years for each crop
df['Total_Area'] = df[['Area 2006-07', 'Area 2007-08', 'Area 2008-09', 'Area 2009-10', 'Area 2010-11']].sum(axis=1)

# Plotting the histogram
plt.figure(figsize=(10, 6))
sns.histplot(df['Total_Area'], bins=20, kde=True, color='green')
plt.xlabel('Total Area (2006-2011)', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.title('Distribution of Total Area (2006-2011) for Crops', fontsize=17)
plt.grid(True)
plt.show()
```
