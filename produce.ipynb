```python
import pandas as pd

# Assuming your DataFrame is named df
missing_data = df.isna().sum()
missing_percentage = (missing_data / len(df)) * 100
print(missing_percentage)
```

```python
# Drop columns with more than 50% missing data
threshold = 50  # You can adjust this threshold
columns_to_drop = missing_percentage[missing_percentage > threshold].index
df = df.drop(columns=columns_to_drop)

# Alternatively, drop rows with missing data
df = df.dropna()
```

```python
# Separate numeric and non-numeric columns
numeric_cols = df.select_dtypes(include=['number']).columns
non_numeric_cols = df.select_dtypes(exclude=['number']).columns

print("Numeric columns:", numeric_cols)
print("Non-numeric columns:", non_numeric_cols)
```

```python
# Fill missing values in numeric columns with median
df[numeric_cols] = df[numeric_cols].apply(lambda x: x.fillna(x.median()), axis=0)
```

```python
# For example, fill missing values in non-numeric columns with a placeholder
df[non_numeric_cols] = df[non_numeric_cols].fillna('Unknown')
```

```python
# Verify that there are no missing values left
print(df.isna().sum())
```

```python
# Describe the data
print(df.describe())
```

```python
# Plot the data
fig, ax = plt.subplots(figsize=(22, 11))  # Increase the size of the figure

df.plot(kind='line', x='Particulars', y=df.columns[3:], ax=ax)

# Rotate x-axis labels for better readability
plt.xticks(rotation=30, ha='right', fontsize=12)  # Rotate labels and set font size

# Set the title and labels for clarity
plt.title('Yearly Data Analysis', fontsize=20)
plt.xlabel('Particulars', fontsize=16)
plt.ylabel('Values', fontsize=16)

# Show grid for better readability
plt.grid(True)

# Show the plot
plt.show()
```

```python
# Plot the heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(df.iloc[:, 3:].corr(), annot=True, cmap='coolwarm', fmt=".2f")

# Set the title and labels
plt.title('Correlation Heatmap', fontsize=16)
plt.xlabel('Variables', fontsize=12)
plt.ylabel('Variables', fontsize=12)

# Adjust layout to prevent cutoff of labels
plt.tight_layout()

# Show plot
plt.show()
```
