```python
# Check the exact column names
print("Column Names:", df.columns)

# Standardize column names by stripping leading/trailing spaces
df.columns = df.columns.str.strip()

# Verify the corrected column names
print("Corrected Column Names:", df.columns)
```

```python
# Use the corrected column name for pivot table
pivot_data = df.pivot_table(index='State', columns='Crop', values='Yield (Quintal/ Hectare)', aggfunc='sum')

# Iterate over each state to create a pie chart for each one
states = pivot_data.index

for state in states:
    state_data = pivot_data.loc[state].dropna()  # Drop NaN values
    plt.figure(figsize=(8, 6))
    wedges, _, autotexts = plt.pie(state_data, autopct='%1.1f%%', startangle=140)
    
    # Improve the visualization with colors and legend
    for autotext in autotexts:
        autotext.set_fontsize(12)
        autotext.set_color('white')
    
    plt.title(f'Crop Yield Distribution in {state}', fontsize=16)
    plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
    plt.legend(wedges, state_data.index, title="Crops", loc="center left", bbox_to_anchor=(1, 0, 0.5, 1))
    plt.show()
```

```python
# Bar Chart of Average Yield per Crop

average_yield_per_crop = df.groupby('Crop')['Yield (Quintal/ Hectare)'].mean()

plt.figure(figsize=(10, 6))
average_yield_per_crop.sort_values().plot(kind='barh', color='skyblue')
plt.xlabel('Average Yield (Quintal/ Hectare)')
plt.ylabel('Crop')
plt.title('Average Yield per Crop Across States')
plt.show()
```

```python
# Box Plot of Yield Distribution per Crop
plt.figure(figsize=(12, 8))
df.boxplot(column='Yield (Quintal/ Hectare)', by='Crop', grid=False, vert=False)
plt.xlabel('Yield (Quintal/ Hectare)')
plt.title('Yield Distribution per Crop')
plt.suptitle('')  # Suppress the automatic title to only show custom title
plt.show()
```

```python
# Heatmap of Crop Yields by State
pivot_data = df.pivot_table(index='State', columns='Crop', values='Yield (Quintal/ Hectare)', aggfunc='mean')

plt.figure(figsize=(14, 10))
sns.heatmap(pivot_data, annot=True, fmt=".1f", cmap="YlGnBu")
plt.title('Heatmap of Crop Yields by State')
plt.xlabel('Crop')
plt.ylabel('State')
plt.show()
```

```python
# Stacked Bar Chart of Total Yield per State
total_yield_per_state = df.pivot_table(index='State', columns='Crop', values='Yield (Quintal/ Hectare)', aggfunc='sum', fill_value=0)

total_yield_per_state.plot(kind='bar', stacked=True, figsize=(15, 9))
plt.xlabel('State')
plt.ylabel('Total Yield (Quintal)')
plt.title('Total Yield per State by Crop')
plt.legend(title='Crop', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```
